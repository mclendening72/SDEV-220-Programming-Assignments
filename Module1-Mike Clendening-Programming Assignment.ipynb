```python
sec_per_min = 60
min_per_hour = 60
hours_per_day = 24

secs_per_hour = sec_per_min * min_per_hour
secs_per_day = sec_per_hour * hours_per_day

print ("There are", secs_per_hour, "seconds per hour")
print ("There are", secs_per_day, "seconds per day")

result1 = secs_per_day / secs_per_hour
print (result1)

result2 = secs_per_day // secs_per_hour
print (result2)

# The number did agree with the exception on the .0
```
